```python
# Train a model
import src.gesture_model as gm
import pandas as pd
import src.crud as crud
import pickle

data_path = "data/"
model_path = "models/"

# load the data and gesture-index map
data = crud.load_csvs(data_path = data_path)
gesture_index_map = crud.load_json_mapping(
    "configs/gesture_index_map.json"
)

# train the model
model = gm.train_gesture_model(
    data = data,
    gesture_index_map = gesture_index_map,
    sample_size_per_gesture = 100,
    sample_size_other_gesture = 300)

# pickle the model
pickle.dump(model, model_path + "gesture_model.pkl")
    



```
